# Query aware reps VS query aware updates
* EGCN, GSPR, HDE all use query aware node representations. This means that the passage-based contextual-embedding of each entity is combined with the global query representation. Meaning each node is aware of both its meaning in the text, but also the query.

* An alternative to forcing both the query and entity information into a single vec is to instead incorporate the query into the node state update process. In this way node states need not explicitly retain query info, instead the query is factored into the gate/attention/linear transformations involved in updating the node states

# Attentional updates
* both EGCN and GSPR use a sum over neighbours states to get the update message for a given node. this sum is used either in a gate, or simple transformation to determine the nodes next state. EGCN transforms the nighbour states before summing while GSPR sums before transforming. 

* while it is true that transforming before summing may allow the learned transformation to ignore unimportant information - this affect would be limited. Instead a much more explict attention like mechanism can be used such as the one described in GAT. Doing this replaces the sum over neighbour states with a weighted sum where the scalar weightings are decided by a learned attention function. This allows the attention structure to explicitly and selectively ignore information from certain nodes, while boosting the influence of others
