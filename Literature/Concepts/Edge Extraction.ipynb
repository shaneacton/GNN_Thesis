# Coreference

When thinking about entities and relationships found in natural language text, it is usefulto know which entity mentions are refering to the same logical entity. The trivial case is where the same named entity label is used eg: "Germany is in Europe. Berlin is the capitl of Germany" here the label 'Germany' is used twice to refer to the same entity.

Coreferences are generally secondary labels used to refer to already mentioned entities eg:"I put the tablecloth on the table to protect it" . Here 'it' refers back to 'the table'. Older heuristic based coreference resolvers can struggle to deal with ambiguous coreferences which require a nuanced understanding of the text. For example in the above example, without knowing anything about the entities table and tablecloth and their natural relationship, a system may mistakenly think the 'it' refers to 'the tablecloth' instead of 'the table'.

More recently NN based coreference resolvers which are able to leverage a semmantic understanding of words and their relationships have become SOTA
