* 3 types of outputs discussed - node selection, where a confidence value is outputted for every node, which are then all softmaxed. Graph classification can produce a general output which is a function of the entire graph. And sequence generation


* To handle graph-level classifications, they suggest to create a dummy super node that is connected to all other nodes by a special type of edge. Thus, graph-level regression or classification can be handled in the same manner as node-level regression or classification


* Sequence generation can be built on either node selection or graph classification, in either case, an output is produced for each of a specified number of output steps. The output function and the function which carries state over from one output step to the next can be two separate GNN's, which propagate information around nodes separately and interact only indirectly. The two functions can also share a propagation process whereby the inside of each output step, one propagation process is completed, after which the output function and the state propagation function both produce a outputs based on the same node states. The stopping condition can be a fixed number of output elements, or some stopping condition eg: when the model outputs a full-stop.
