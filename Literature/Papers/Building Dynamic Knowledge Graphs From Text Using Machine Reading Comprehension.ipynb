# Notes
* paper concerns itself with constructing knowledge graphs for information which changes updates over time. In many domains, the nature of entities and their relationships evolve, such as in News and in proceedural text (sequential instructions) eg cooking guides.
* Is able to change entities and relations as it encounters new information.
* In contrast, a static KG would only add in new information, potentially losing information about the time ordering of relationship transformations. Some inconsistencies which may arise include eggs which are simultaneously fresh, and cooked in a static KG.
* A dynamic KG would record the eggs as being fresh, then re-evalute this when new information is encountered about the eggs having been cooked.
* The graph update function uses attention, such that entities can be updated in a near null way when newly encountered info is irrelevant to them.
* While the primary goal of the paper is to construct knowledge graphs, the system is targetted at maximising the performance of MRC tasks. As a part of trying to solve MC, the system inadvertently learns to construct semantically rich and accuracte knowledge graphs. This illustrates the usefulness of downstream goals in training such systems. There may yet be more to be gained from diversifying and generalising the types of goals that such a system is targetted on - ie, training the same system using differing datasets which require differing skills to solve. Only high quality and general knowlege graphs would be useful in solving differing tasks simultaneously. An example of such related but differing tasks may include: next-word or next-sentence prediction, Contradiction detection, passage summarisation, or any other sufficiently similar NLP tasks
* the paper appears to be highly non-general, focusing very strongly on locations of entities, ie which entities contain which other entities. Thus some of the solutions the paper proposes may be unadaptable to more general systems which keep track of many types of entity properties



# Datasets
* easy - bAbI dataset (Weston et al., 2015) includes questions about movement of entities; however, its language is generated synthetically over a small lexicon, and hence models trained on bAbI often do not generalize well when tested on real-world data
* PROREAD (Berant et al., 2014) introduced the PROCESSBANK dataset, which contains paragraphs of procedural text as in PROPARA. However, this earlier task involves mining arguments and rela2 tions from events, not tracking the dynamic state changes of entities
* main -  PROPARA dataset (Dalvi et al., 2018), which comprises procedural text about scientific processes
